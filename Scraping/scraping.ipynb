# Парсинг сайта. Харлов МИТ-101

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib
def openurl(url):
    try:
        html=urlopen(url,timeout=5)
    except urllib.error.HTTPError as e:
        return None
    except Exception:
        return None
    except  urllib.error.URLError as b:
        return None
    return html
```

```python
html = openurl("https://openedu.ru/course/")
code=BeautifulSoup(html.read().decode('utf-8'))
```

```python
import re
import json
import csv
```

```python
scripts=code.findAll("script",{ "type":"text/javascript"})
```

```python
s=scripts[12].get_text()
```

# Ищем ссылки на все курсы

```python
a=re.findall(r'"url": "(.*?)"', s)
newhref=[]
for i in range(len(a)):
    newhref.append("https://openedu.ru"+a[i])
```

# Основной код
ПРоходим по каждой ссылке и ищем по нему информацию.

```python
file=open("courses.csv", "a", newline="",encoding='utf-8')
columns=["title","description","len","teacher"]
dictBig={}
for i in range(10):
    inf={"title":"","description":"","lenght":"","teacher":""}
    html=openurl(newhref[i])
    if html!=None:
        try:
            curs=BeautifulSoup(html.read().decode('utf-8'))
        except Exception:
            continue
        inf["title"]=curs.h1.get_text()
        lect=[]
        inf["lenght"]=curs.find("ul",{"class":"list-unstyled"}).find("li",{"class":"clearfix"}).get_text().split("            ")[1][:-1]
        for v in curs.findAll("div",{"class":"card card-lecturer"}):
            lect.append(v.find("h3").get_text())
        inf["teacher"]=lect
        desc=""
        ds=curs.find("div",{"class":"description"})
        if ds!=None:
            for a in ds.findAll("p"):
                desc+=a.get_text()
            inf["description"]=desc
        else:
            continue
        dictBig[newhref[i]]=inf
        print(inf["title"])
    else:
        continue
```
